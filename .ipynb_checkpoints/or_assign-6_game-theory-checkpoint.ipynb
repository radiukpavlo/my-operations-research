## Дано

Нехай маємо математичну модель прямої задачі лінійного програмування:

$$\begin{array}{rcl} f = \min\limits_{t \in T} \{ t_{1} + t_{2} + t_{3} \},
\end{array}$$

$$\begin{cases}
\begin{aligned}
7t_{1} + 2t_{2} + 3t_{3} \geq 1; \\
3t_{1} + 4t_{2} + 5t_{3} \geq 1; \\
2t_{1} + 6t_{2} + 4t_{3} \geq 1; \\
5t_{1} + 4t_{2} + 6t_{3} \geq 1; \\
\end{aligned}
\end{cases}$$

$$\begin{array}
1t_{i} \geq 0, \quad i = \overline{1,3}.
\end{array}$$

## Блок 0. Підключення необхідних пакетів

````python
import warnings
warnings.filterwarnings('ignore')

from pulp import *
import matplotlib.pyplot as plt
import numpy as np
````

## Блок 1. Пряма задача лінійного програмування probDirect

### Блок 1.1. Реалізація прямої задачі лінійного програмування

````python
# Створення об'єкту математичної моделі
probDirect = LpProblem("Пряма задача лінійного програмування", LpMinimize)
````

````python
# Оголошення змінних мат. моделі
t1 = LpVariable(name="t1", lowBound=0, upBound=None) # оголошення змінної x1 >= 0
t2 = LpVariable(name="t2", lowBound=0, upBound=None) # оголошення змінної x2 >= 0
t3 = LpVariable(name="t3", lowBound=0, upBound=None) # оголошення змінної x3 >= 0
````

````python
# Оголошення та ініціалізація цільової функції
probDirect += t1 + t2 + t3
````

````python
# Оголошення та ініціалізація обмежень моделі
probDirect += 7*t1 + 2*t2 + 3*t3 >= 1, "1st const"
probDirect += 3*t1 + 4*t2 + 5*t3 >= 1, "2nd const"
probDirect += 2*t1 + 6*t2 + 4*t3 >= 1, "3nd const"
probDirect += 5*t1 + 4*t2 + 6*t3 >= 1, "4th const"
````

````python
# Відображення математичної моделі
probDirect
````

````python
# Розв'язок ЗЛП
probDirect.solve(PULP_CBC_CMD(msg=1))
````

### Блок 1.2. Виведення результату розв'язку прямої задачі лінійного програмування

````python
# Виведення результату оптимізації
print ("Статуc ЗЛП: ", LpStatus[probDirect.status])
````

````python
print("Результат:")
for v in probDirect.variables():
    print (v.name, "=", v.varValue)
````

````python
print ("Оптимальне значення цільової функції: f =",
       value(probDirect.objective))
````

## Блок 2. Двоїста задача лінійного програмування

Складемо математична модель до двоїстої задачі лінійного програмування:

$$\begin{array}{rcl} z = \max\limits_{u \in U} \{ u_{1} + u_{2} + u_{3} + u_{4} \},
\end{array}$$

$$\begin{cases}
\begin{aligned}
7u_{1} + 3u_{2} + 2u_{3} + 5u_{4} \leq 1; \\
2u_{1} + 4u_{2} + 6u_{3} + 4u_{4} \leq 1; \\
3u_{1} + 5u_{2} + 4u_{3} + 6u_{4} \leq 1; \\
\end{aligned}
\end{cases}$$

$$\begin{array}
1u_{j} \geq 0, \quad i = \overline{1,4}.
\end{array}$$

### Блок 2.1. Реалізація двоїстої задачі лінійного програмування

````python
# Створення об'єкту математичної моделі
probDual = LpProblem("Двоїста задача лінійного програмування", LpMaximize)
````

````python
# Оголошення змінних мат. моделі
u1 = LpVariable(name="u1", lowBound=0, upBound=None)
u2 = LpVariable(name="u2", lowBound=0, upBound=None)
u3 = LpVariable(name="u3", lowBound=0, upBound=None)
u4 = LpVariable(name="u4", lowBound=0, upBound=None)
````

````python
# Оголошення та ініціалізація цільової функції
probDual += u1 + u2 + u3 + u4
````

````python
# Оголошення та ініціалізація обмежень моделі
probDual += 7*u1 + 3*u2 + 2*u3 + 5*u4 <= 1, "1st const"
probDual += 2*u1 + 4*u2 + 6*u3 + 4*u4 <= 1, "2nd const"
probDual += 3*u1 + 5*u2 + 4*u3 + 6*u4 <= 1, "3nd const"
````

````python
# Відображення математичної моделі
probDual
````

````python
# Розв'язок ЗЛП
probDual.solve(PULP_CBC_CMD(msg=1))
````

### Блок 2.2. Виведення результату розв'язку двоїстої задачі лінійного програмування

````python
# Виведення результату оптимізації
print ("Статуc ЗЛП: ", LpStatus[probDual.status])
````

````python
print("Результат:")
for v in probDual.variables():
    print (v.name, "=", v.varValue)
````

````python
print ("Оптимальне значення цільової функції: z =",
       value(probDual.objective))
````
